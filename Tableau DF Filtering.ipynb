```
import pandas as pd
tableau_df = pd.read_csv("tableau_ready_colleges.csv")
tableau_df
```

```
numerical = tableau_df.select_dtypes(include=[int, float])
objects = tableau_df.select_dtypes(include=[object])
```

```
numerical['Institution Name'] = tableau_df['Institution Name']
```

```
objects = objects.drop_duplicates(subset=['Institution Name'], keep='first').reset_index()
objects
```

```
numerical = numerical.groupby('Institution Name').agg('median').reset_index()
numerical
```

```
final_df = pd.merge(left=objects, right=numerical, on="Institution Name").drop(columns='index')
final_df
```

```
final_df.to_csv("tableau_ready_colleges_final.csv")
```
